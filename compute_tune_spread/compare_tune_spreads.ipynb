### Compare the tune spread with amplitude obtained from the PyEHADTAIL simulation adn the analytical calculations

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

#### Plotting parameters

```python
params = {'legend.fontsize': 20,
          'figure.figsize': (9.5, 8.5),
          'axes.labelsize': 27,
          'axes.titlesize': 23,
          'xtick.labelsize': 27,
          'ytick.labelsize': 27,
          'image.cmap': 'jet',
          'lines.linewidth': 3,
          'lines.markersize': 10,
          'font.family': 'sans-serif'}

plt.rc('text', usetex=False)
plt.rc('font', family='serif')
plt.rcParams.update(params)
```

#### Load the files

```python
set1 = pickle.load(open('./output/DQy_vs_ayy_analytical_appx179.3585107_apxy0.0_2.pkl', 'rb'))
set2 = pickle.load(open('./output/DQy_vs_ayy_xy_new.pkl', 'rb'))
```

```python
plt.plot(set1['ayy (1/m)'], set1['DQy'], '.-', label='analytical formula' )
plt.plot(set2['ayy (1/m)'], set2['DQy'], '.-', label='PyHEATDAIL')
plt.grid()
plt.legend()
plt.xlabel('ayy (1/m)')
plt.ylabel(r'$\Delta Qy$')
savefig = False
if savefig:
    plt.savefig('DQ_pyheadtail_vs_analytical.png')
```
