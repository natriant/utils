### Compare the footprint as obtained from two different simulation codes.
- Sixtracklib vs PyHEADTAIL
- The data are in pickle files produced from the following scripts:
    - PyHEADTAIL_footrpint_test_detuning_coefficients.ipynb
    - Sixtracklib_footrpint.ipynb.ipynb

```python
import pickle
import numpy as np
from math import *
import NAFFlib as pnf
import matplotlib.pyplot as plt
```

### Plotting parameters

```python
params = {'legend.fontsize': 20,
          'figure.figsize': (9.5, 8.5),
          'axes.labelsize': 27,
          'axes.titlesize': 23,
          'xtick.labelsize': 27,
          'ytick.labelsize': 27,
          'image.cmap': 'jet',
          'lines.linewidth': 1,
          'lines.markersize': 7,
          'font.family': 'sans-serif'}

plt.rc('text', usetex=False)
plt.rc('font', family='serif')
plt.rcParams.update(params)
```

```python
tunes_sixtracklib = pickle.load(open('tunes_sixtracklib.pkl', 'rb'))
tunes_pyheadtail = pickle.load(open('tunes_pyheadtail.pkl', 'rb'))
```

```python
plt.plot(tunes_pyheadtail['Qx'], tunes_pyheadtail['Qy'], '.', label='PyHEADTAIL')
plt.plot(tunes_sixtracklib['Qx'], tunes_sixtracklib['Qy'], '.', label='Sixtracklib')
plt.ylim(0.1797, 0.1797+0.0004)
plt.xlim(0.128, 0.128+0.004)
plt.xlabel('Qx')
plt.ylabel('Qy')
plt.grid()
plt.title('900 macroparticles')
plt.legend()
savefig = False
if savefig:
    plt.savefig('footprint_sixtracklib_vs_pyheadtail.png')
```
