#### Plot the total power for different rms ksi required to obtain a PSD of 2.3e-21 rad^2/Hz at the betatron frequency.

The values are computed using the script job002, by inserting manually different phi_0 values until the required PSD is obtained (not a very scientific way to do this but it does its job).

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Function for creating the noise signal

```python
std_list = [0.04, 0.06, 0.08, 0.2, 0.4, 0.6, 0.8]
total_power_list = [3.27e-18, 7.212e-18, 1.3e-17, 6.143e-17, 1.068e-16, 1.065e-16, 1.045e-16]
```

```python
plt.plot(std_list, total_power_list, c='k')
for index, std in enumerate(std_list):
    plt.plot(std, total_power_list[index], '.')


#plt.plot(std_list, total_power_list, c='k')

plt.xlabel('rms('+r'$\xi$'+')'+r'$/2\pi$')
plt.ylabel('Total noise power'+ '(rad'+r'$^2$'+')')
plt.grid(linestyle='--')
#plt.savefig('total_noise_power_vs_rmsKsi_v2.png')
```
