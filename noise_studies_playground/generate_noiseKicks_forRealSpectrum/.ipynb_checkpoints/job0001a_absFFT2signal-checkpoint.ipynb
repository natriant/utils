### Create timeseries from given absolute values of an FFT when the initial timeseries is known


To create a timeseries from the absolute values of an FFT we apply an inverse Fourier transform.

However, the absoulte values of the FFT don't have any information about its phase.
That's why we keep in a variabe, theta, the angles, which we multiply with the absolute values of the FFT.


**Important**:

1) The output of the numpy.fft.fft has the following form: $x+iy$

2) For a complex number: $x + jy = r e^{j \theta}$, where r is the magnitude at a given frequency and $\theta$ is the angle. 

For better explanation, with figures included, see here: https://stackoverflow.com/questions/28618591/numpy-fft-what-is-the-return-value-amplitude-phase-shift-or-angle

3) $r=\sqrt{x^2+y^2}$ = np.abs(fft) and $\theta = \arctan{\frac{y}{x}}$


**The process that is followed is the following:**
1) Create a signal in the time domain, $signal$

2) Move to frequency domain, by applying FFT

3) Compute the amplitude of the FFT, $r$, and the phase, $theta$

4) Reconstruct the FFT by the amplitude and the phase information, $fft_2$

5) Recover the original signal by applying inverse fourier transform, $signal _reconstructed$

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Create signal in time domain

```python
T = 0.5 # s
N = 10000 # 
f = 10 # Hz
```

```python
t = np.linspace(0, T, N)
Dt = t[1]-t[0] # Delta t
```

```python
signal = np.sin(2*np.pi*f*t)
```

```python
plt.plot(t, signal)
plt.xlabel('time (s)')
plt.ylabel('Amplitude (a.u.)')
```

### 2. FFT --> Frequency domain

```python
fft = np.fft.fft(signal)
freq = np.fft.fftfreq(N, Dt)
```

```python
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(fft))
plt.xlim(-20, 20)
plt.xlabel('Hz')
plt.ylabel('Amplitude (a.u.)')
```

### 3. Compute the amplitude of the FFT, $r$, and the phase, $theta$

```python
r = np.sqrt(np.real(fft)**2+np.imag(fft)**2) # = np.abs(fft)
theta = np.arctan2(np.imag(fft), np.real(fft))
```

### 4. Reconstruct the FFT by the amplitude and the phase information

```python
fft_2 = r*np.exp(1j*theta)
```

### 5. Recover the original signal by applying inverse fourier transform

```python
signal_reconstructed = np.fft.ifft(fft_2)
```

### 6. Compare the orginal signal with the recovered one

```python
plt.plot(t, signal, linewidth = 5, label = 'original')
plt.plot(t, signal_reconstructed, linestyle='--', label='reconstructed')
plt.xlabel('time (s)')
plt.ylabel('Amplitude (a.u.)')
plt.legend(loc=1)
```
